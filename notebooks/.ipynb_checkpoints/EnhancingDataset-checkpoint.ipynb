```python
link_data = '../datasets/ml-latest-small/links.csv'
df_links = pd.read_csv(link_data, low_memory=False)
poster_data = '../datasets/movie_poster/MovieGenre.csv'
df_posters = pd.read_csv(poster_data, low_memory=False, encoding='latin-1')
youtube_data = '../datasets/ml-youtube.csv'
df_youtube = pd.read_csv(youtube_data, low_memory=False)


df_movies['imdbId'] = df_links['imdbId']
df_movies['imdbLink'] = 'None'
df_movies['poster'] = 'None'
df_movies['youtubeId'] = 'None'


for index, row in df_movies.iterrows():
    if not df_posters[df_posters['imdbId']==row.imdbId].empty:
        idx_df_posters = int(df_posters[df_posters['imdbId']==row.imdbId].index[0])
        idx_df_movies = int(df_movies[df_movies['imdbId']==row.imdbId].index[0])

        df_movies.loc[idx_df_movies,('imdbLink')] = df_posters.loc[idx_df_posters,('Imdb Link')]
        df_movies.loc[idx_df_movies,('poster')] = df_posters.loc[idx_df_posters,('Poster')]
    
    
    if not df_youtube[df_youtube['movieId']==row.movieId].empty:
        idx_df_youtube = int(df_youtube[df_youtube['movieId']==row.movieId].index[0])
        idx_df_movies = int(df_movies[df_movies['movieId']==row.movieId].index[0])

        df_movies.loc[idx_df_movies,('youtubeId')] = df_youtube.loc[idx_df_youtube,('youtubeId')]


df_movies.head(10)
```

```python
df_movies.to_csv('../datasets/enhanced_movies.csv', index=False)
```

```python
enhanced_movies_data = '../datasets/enhanced_movies.csv'
df_enhanced_movies = pd.read_csv(enhanced_movies_data, low_memory=False)
df_enhanced_movies.head()
```

```python
grouped_tags_data = '../datasets/grouped_tags.csv'
df_grouped_tags = pd.read_csv(grouped_tags_data, low_memory=False)
df_grouped_tags.head()


df_enhanced_movies['tags'] = 'None'

for index, row in df_enhanced_movies.iterrows():
    
    if not df_grouped_tags[df_grouped_tags['movieId']==row.movieId].empty:
        
        idx_grouped_tags = int(df_grouped_tags[df_grouped_tags['movieId']==row.movieId].index[0])
        idx_enhanced_movies = int(df_enhanced_movies[df_enhanced_movies['movieId']==row.movieId].index[0])

        df_enhanced_movies.loc[idx_enhanced_movies,('tags')] = df_grouped_tags.loc[idx_grouped_tags,('tags')]
        
df_enhanced_movies.tail()
```

```python
my_movies = '../datasets/enhanced_movies_tags.csv'
df_my_movies = pd.read_csv(my_movies, low_memory=False)
df_my_movies.head()
```

```python
def read_as_json(file_name):
    with open(file_name) as json_file:
        data = json.load(json_file)
    return data
```

```python
descrip = read_as_json('Descritive_Analyze/all.json')
print(descrip[0])
```

```python
df_descript = df_G1 = pd.DataFrame.from_dict(descrip)
df_descript.head()
```

```python
df_my_movies['r_counter'] = 'None'

for index, row in df_my_movies.iterrows():
   
    idx_df_descript = int(df_descript[df_descript['movieId']==row.movieId].index[0])
    idx_df_my_movies = int(df_my_movies[df_my_movies['movieId']==row.movieId].index[0])

    df_my_movies.loc[idx_df_my_movies,('youtubeId')] = df_descript.loc[idx_df_descript,('youtubeId')]
    
    print(row.movieId)
```

```python
"""
Serão definidos 5 Grupos de filmes, tal que:

    G1, filmes bastante avaliados, tq, ratings(G1) > 100
    G2, filmes muito avaliados, tq, 100 >= ratings(G2) > 50
    G3, filmes razoavelmente avaliados, tq, 50 >= ratings(G3) > 20
    G4, filmes pouco avaliados, tq, 20 >= ratings(G4) > 5
    G5, filmes pouqíssimo avaliados, tq, 5 >= ratings(G5)
"""

G1 = []
G2 = []
G3 = []
G4 = []
G5 = []

for item in descrip:
    if item['counter'] > 100:
        G1.append(item)
    elif 100 >= item['counter'] > 50:
        G2.append(item)
    elif 50 >= item['counter'] > 20:
        G3.append(item)
    elif 20 >= item['counter'] > 5:
        G4.append(item)
    else:
        G5.append(item)
        
print("Tamanho dos grupos: G1={}, G2={}, G3={}, G4={} e G5={}".format(len(G1), len(G2), len(G3), len(G4), len(G5)))
```
